```python
import os
import pandas as pd
import numpy as np
from shutil import copyfile
from sklearn.model_selection import train_test_split
```

```python
# Init a path var for the dataset
path_global = "dataset/Images/"
```

Let's see the different categories that we have

```python
categories = []
for i in os.listdir(path_global):
    categories.append(i)
```

# Create a DataFrame

```python
categ = []
path = []
name = []
for i in os.listdir(path_global):
    for j in os.listdir(path_global + i):
        categ.append(i)
        name.append(j)
        path_ = path_global + i + "/" + j
        path.append(path_)
        
data = pd.DataFrame(categ, columns=['Categ'])
data['Name'] = name
data['Path'] = path
data['Categ_Cleaned'] = [i.split('-')[1] for i in data['Categ']]
```

```python
data.head() 
```

# Create train, validation and test dataset

```python
x = data.copy()
y = data[['Categ_Cleaned']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(
    x_train, y_train, test_size=0.25, random_state=42)
```

# Create new folders for Train, Validation, and Test dataset

```python
def createDirectory(categ, direct=['Train', 'Validation', 'Test']):
    for cat in categ:
        for d in direct:
            if os.path.isdir("dataset/" + d + "/" + cat):
                print("The directory :", "data/" + d + "/" + cat, "was already created")
            else:
                os.makedirs("dataset/" + d + "/" + cat)
                print("The directory :", "data/" + d + "/" + cat, "was created successfully")
```

```python
createDirectory(categories)
```

# Put images in the right folder

```python
def moveImages():
    path_train = []
    path_test = []
    path_val = []
    for index in x_train.index:
        try:
            cat = x_train.loc[index, 'Categ']
            path_ = "dataset/Train/" + cat.strip() + "/" + x_train.loc[index, 'Name']
            copyfile("dataset/Images/" + cat.strip() + "/" + x_train.loc[index, 'Name'], path_)
            path_train.append(path_)
        except:
            print("Picture not in the path")
            path_ = " "
            path_train.append(path_)
            pass
    path_train = pd.DataFrame(path_train, index=x_train.index, columns=['newPath'])

    for index in x_val.index:
        try:
            cat = x_val.loc[index, 'Categ'] 
            path_ = "dataset/Validation/" + cat.strip() + "/" + x_val.loc[index, 'Name']
            copyfile("dataset/Images/" + cat.strip() + "/" + x_val.loc[index, 'Name'], path_)
            path_val.append(path_)
        except:
            print("Picture not in the path")
            path_ = " "
            path_val.append(path_)
            pass
    path_val = pd.DataFrame(path_val, index=x_val.index, columns=['newPath'])

    for index in x_test.index:
        try:
            cat = x_test.loc[index, 'Categ'] 
            path_ = "dataset/Test/" + cat.strip() + "/" + x_test.loc[index, 'Name']
            copyfile("dataset/Images/" + cat.strip() + "/" + x_test.loc[index, 'Name'], path_)
            path_test.append(path_)
        except:
            print("Picture not in the path")
            path_ = " "
            path_test.append(path_)
            pass
    path_test = pd.DataFrame(path_test, index=x_test.index, columns=['newPath'])
    
    return path_train, path_val, path_test
```

```python
cat = x_train.loc[10535, 'Categ'].strip()
"dataset/Images/" + cat.strip() + "/" + x_train.loc[10535, 'Name']
```

```python
path_train, path_val, path_test = moveImages()
```

```python
x_train = pd.merge(x_train, path_train, left_index=True, right_index=True, how='inner')
x_val = pd.merge(x_val, path_val, left_index=True, right_index=True, how='inner')
x_test = pd.merge(x_test, path_test, left_index=True, right_index=True, how='inner')
```

Concat to get only one file with all train validation and test values

```python
x_train['Label'] = 'Train'
x_val['Label'] = 'Validation'
x_test['Label'] = 'Test'

y_train['Label'] = 'Train'
y_val['Label'] = 'Validation'
y_test['Label'] = 'Test'

data = pd.concat([x_train, x_val, x_test])
target = pd.concat([y_train, y_val, y_test])
```

```python
data.to_csv('dataset/data.csv')
target.to_csv('dataset/target.csv')
```
